```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

```python
iris=load_iris()
```

```python
x=iris.data
y=iris.target
x
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=45)
```

```python
from sklearn.neighbors import KNeighborsClassifier as KNN
knn=KNN(n_neighbors=3)
knn.fit(x_train,y_train)
```

pickle.dump to serialize an object hierarchy, you simply use dump()

pickle.load to deserialize a data stream, you call the loads() function.

```python
import pickle
# save the trained model as a pickle string.
saved_model=pickle.dumps(knn)
# Load the pickled model

knn_from_pickle=pickle.loads(saved_model)
```

```python
# Use the Loaded pickled model to make predictions

knn_from_pickle.predict(x_test)
```

```python
import numpy as np
s=np.array([4.8, 3.4, 1.6, 0.2])
#newdataofnewpatient=np.array([15.50,14.05,125.10,765.1,0.09847,0.11570,0.098750,0.079530,0.1739])
s
s=s.reshape(1,-1)
#setosa
knn_from_pickle.predict(s)
```

Pickled model as a file using joblib.

Joblib is the replacement of pickle as it is more efficent on objects that carry large numpy arrays. These functions also
accept file-like object instead of file names.

joblib.dump to serialize an jobject hierarchy


joblib.load to deserialize a data stream

```python
from sklearn.externals import joblib

# save the model as a pickle in a file

joblib.dump(knn,'filename.pkl')
# load the model from the file

knn_from_joblib=joblib.load('filename.pkl')

#use the loaded model to make predictions

knn_from_joblib.predict(x_test)
```
